```python
#!/usr/bin/env python

# (1)
# Write a function named uses_only that takes a word and a string of letters,
# and that returns True if the word contains only letters in the list.
#   - write uses_only
# (2)
# Can you make a sentence using only the letters acefhlo? Other than "Hoe
# alfalfa?"
#   - write function to assist you
#   - manually type favorite sentence(s) here:
#       1: cocoa coffee heal
#       2: aloe heal face
#       3: local elf hello 
##############################################################################
# Imports
import os

# Body
def make_list():
    """creates a list with words within the text file"""
    lst = []
    txt = input('Insert the name of the text file (example: words.txt): ')
    file_handle_words = open(txt, 'r')
    for word in file_handle_words:
        lst.append(word.split('\n')[0])
    return lst

def uses_only(word, onlyFromHere):
    flag = False
    for letter in word:
        for letter_OFH in onlyFromHere:
            if letter != letter_OFH:
                flag = False
            else:
                flag = True
                break
        if(flag == False):
            break
    return flag

def make_word():
    lst = make_list()
    for word in lst: 
        if (uses_only(word, 'acefhlo')):
            print(word)
    

##############################################################################
def main():
    make_word()

if __name__ == '__main__':
    main()
```
