```python
#!/usr/bin/env python3

# (1)
# Write a function named avoids that takes a word and a string of forbidden
# letters, and that returns True if the word doesn't use any of the forbidden
# letters.
#   - write avoids
# (2)
# Modify your program to prompt the user to enter a string of forbidden
# letters and then print the number of words that don't contain any of them.
#   - write forbidden_prompt and
#   - modify to create forbidden_param that accepts the string as an argument
# (3)
# Can you find a combination of 5 forbidden letters that excludes the smallest
# number of words?
#   - write a function that finds this combination of letters: find_five
#   - have that function print the letters and print the # of words excluded
##############################################################################
# Imports
import os

# Body


def make_list():
    """creates a list with words within the text file"""
    lst = []
    txt = input('Insert the name of the text file (example: words.txt): ')
    file_handle_words = open(txt, 'r')
    for word in file_handle_words:
        lst.append(word.split('\n')[0])
    return lst

    
def avoids(word, forbidden):
    """ return True if word NOT forbidden"""
    for forbidden_letter in forbidden:
        for letter in word:
            if forbidden_letter == letter:
                return False 
    return True
    
def forbidden_prompt():
    """ print count of words NOT forbidden by input"""
    count = 0
    forbidden_str = input("Please enter a string of forbidden letters: ")
    for word in make_list():
        if(avoids(word, forbidden_str)):
            count = count + 1
    print('The number of words that contains forbidden letters is:', count)

def forbidden_param(string):
    """ return count of words NOT forbidden by param"""
    count = 0
    for word in make_list():
        if avoids(word, string):
            count = count + 1
    print('The number of words that contains forbidden letters is:', count)
    
def forbidden_param_inverse(string, lst):
    """ return count of words forbidden by param"""
    count = 0
    for word in lst:
        if not avoids(word, string):
            count = count + 1
    
    return count
    
def find_five():
    alphabet_string = 'abcdefghijklmnopqrstuvwxyz'
    lst_from_txt = make_list()
    lst = []
    for letter in alphabet_string:
        lst.append([letter, forbidden_param_inverse(letter, lst_from_txt)])
    lst.sort(key = lambda x: x[1]) 
    lst = lst[0:5]
    print(lst)


##############################################################################
def main():
    find_five()

if __name__ == '__main__':
    main()
```
